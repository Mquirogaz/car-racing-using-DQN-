```python
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python

#CarRacing-v2 con acciones
env = gym.make('CarRacing-v2', continuous=False)
```

```python
#informacion que el agente recibe del entorno. 
print("Espacio de observacion:")
print(env.observation_space)
print("Box(rango de valores de cada pixel, (alto imagen, ancho imagen, canales de color)")
print("")
```

```python
#posibles acciones que el agente (auto) puede tomar(acelerar, girar, etc)
print("Espacio de acciones:")
print(env.action_space)
print("[do nothing, turn left, turn right, accelerate, brake]")
```

```python
s, info = env.reset()
initial_state, info = env.reset()
print("Dimensiones de la imagen inicial:", initial_state.shape)

plt.figure(figsize=(5, 5))
plt.imshow(initial_state)
plt.axis('off')
plt.show()

```

```python
import matplotlib.animation as animation

frames = []
for i in range(50):
    s, r, terminated, truncated, info = env.step(0)
    frames.append(s)


fig = plt.figure(figsize=(5, 5))
plt.axis('off')
im = plt.imshow(frames[0])
def animate(i):
    im.set_array(frames[i])
    return im,
anim = animation.FuncAnimation(fig, animate, frames=len(frames))
HTML(anim.to_jshtml())
```
